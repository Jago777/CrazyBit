```python
import numpy as np
```

```python
def GenChis(h=1,k=0):
    B = []
    for g in range(0,h):
        A = np.random.randint(100)
        B.append(A+k)
    return B
```

```python
from matplotlib import pyplot as plt
```

```python
Demons = ['Dima', 'Grigoriy', 'Taras', 'Velzevul', 'Mefistofel', 'Baron Sub']
KolSoulsInCaptivity = GenChis(len(Demons))
xs = [i + 0.1 for i, _ in enumerate(Demons)]
```

```python
plt.bar(xs, KolSoulsInCaptivity)
plt.ylabel('Количество душ')
plt.title('Великие поработители')
plt.xticks([i + 0.1 for i, _ in enumerate(Demons)], Demons)
plt.show()
```

```python
Geroi = ['John', 'Zatana', 'Raven', 'Demon', 'Angel', 'Sabrina', 'Strange', 'Spider', 'Punisher', 'Blade']
KolPowers = GenChis(len(Geroi))
```

```python
xs = [i + 0.1 for i, _ in enumerate(Geroi)]
plt.bar(xs, KolPowers, 0.7)
plt.ylabel('Количество сил')
plt.title('Могучие герои земли')
plt.xticks([i + 0.1 for i in range(len(Geroi))], Geroi, fontsize = 7)
plt.show()
```
